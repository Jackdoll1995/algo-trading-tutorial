```python
# last 24 hours change statistic
last_24 = client.get_ticker(symbol = 'BTCJPY')
last_24
```

```python
open_time = last_24['openTime']
open_time
```

```python
open_date_time = pd.to_datetime(open_time, unit='ms')
open_date_time
```

```python
close_date_time = pd.to_datetime(last_24['closeTime'], unit='ms')
close_date_time
```

```python
open_price = float(last_24['openPrice'])
open_price
```

```python
high_price = float(last_24['highPrice'])
high_price
```

```python
low_price = float(last_24['lowPrice'])
low_price
```

```python
close_price = float(last_24['lastPrice'])
close_price
```

```python
price_change = close_price - open_price
price_change
```

```python
price_change_percent = price_change / open_price * 100
price_change_percent
```
